# Exploratory data analysis

# Platform

```python
# plotting a bar chart for platform and it's count
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.countplot(x = 'Platform',data = df)


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for platform vs NA_Sales

sales_gen = df.groupby(['Platform'], as_index=False)['NA_Sales'].sum().sort_values(by='NA_Sales', ascending=False)

sns.barplot(x = 'Platform',y= 'NA_Sales' ,data = sales_gen)
```

```python
# plotting a bar chart for platform vs EU_Sales

sales_gen = df.groupby(['Platform'], as_index=False)['EU_Sales'].sum().sort_values(by='EU_Sales', ascending=False)

sns.barplot(x = 'Platform',y= 'EU_Sales' ,data = sales_gen)
```

```python
# plotting a bar chart for platform vs JP_Sales

sales_gen = df.groupby(['Platform'], as_index=False)['JP_Sales'].sum().sort_values(by='JP_Sales', ascending=False)

sns.barplot(x = 'Platform',y= 'JP_Sales' ,data = sales_gen)
```

```python
# plotting a bar chart for platform vs Other_Sales

sales_gen = df.groupby(['Platform'], as_index=False)['Other_Sales'].sum().sort_values(by='Other_Sales', ascending=False)

sns.barplot(x = 'Platform',y= 'Other_Sales' ,data = sales_gen)
```

```python
# plotting a bar chart for platform vs Global_Sales

sales_gen = df.groupby(['Platform'], as_index=False)['Global_Sales'].sum().sort_values(by='Global_Sales', ascending=False)

sns.barplot(x = 'Platform',y= 'Global_Sales' ,data = sales_gen)
```

# Publisher

```python
data_pie = df["Publisher"].value_counts().rename_axis('Publisher').reset_index(name='publisher_count').head(25) 
data_pie
```

```python
#Plotting Pie chart of Publisher and its count
plt.figure(figsize=(20,20))
plt.pie(data_pie.publisher_count, labels=data_pie.Publisher, startangle=90,autopct='%.1f%%')
plt.title('Publisher')
plt.show()
```

```python
# plotting a bar chart for Publisher vs NA_Sales

sales_gen = df.groupby(['Publisher'], as_index=False)['NA_Sales'].sum().sort_values(by='NA_Sales', ascending=False).head(15)

plt.figure(figsize=(20,20))
plt.xticks(rotation=30)
sns.barplot(x = 'Publisher',y= 'NA_Sales' ,data = sales_gen)
```
